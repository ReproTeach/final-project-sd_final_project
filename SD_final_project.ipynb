```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation as LDA
import re
import nltk
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
```

```python
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    # Remove RT (retweet sign)
    text = re.sub(r'rt[\s]+', '', text)
    # Remove mentions
    text = re.sub(r'@\S+', '', text)
    # Remove all non-alphabetic characters
    text = re.sub(r'\W', ' ', text)
    # Remove single characters
    text = re.sub(r'\s+[a-zA-Z]\s+', ' ', text)
    # Remove leading and trailing whitespaces
    text = text.strip()
    # Remove stopwords and do stemming
    # text = " ".join([stemmer.stem(i) for i in text.split() if i not in stop_words])
    return text

merged_df['Text_Cleaned'] = merged_df['OriginalTweet'].apply(preprocess_text)
```

```python
merged_df[['Text_Cleaned','OriginalTweet']]
```

```python
documents = merged_df['Text_Cleaned']
```

```python
count_vectorizer = CountVectorizer(stop_words='english')
```

```python
count_data = count_vectorizer.fit_transform(documents)
```

```python
# Tweak the two parameters below (use int values below 15)
number_topics = 5
number_words = 9

# Create and fit the LDA model
lda = LDA(n_components=number_topics, n_jobs=-1)
lda.fit(count_data)
```

```python
count_vectorizer.get_feature_names_out()
```

```python
words = list(count_vectorizer.get_feature_names_out())
words
```

```python
lda.components_.shape
```

```python
def print_topics(model, count_vectorizer, n_top_words):
    words = count_vectorizer.get_feature_names_out()
    for topic_idx, topic in enumerate(model.components_):
        print("\nTopic #%d:" % (topic_idx+1))
        print(" ".join([words[i]
                        for i in topic.argsort()[:-n_top_words - 1:-1]]))

print_topics(lda, count_vectorizer, number_words)
```

```python
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

# Transform the count data using LDA model
topic_weights = lda.transform(count_data)

# Apply t-SNE for dimensionality reduction
tsne_model = TSNE(n_components=2, random_state=42)
tsne_data = tsne_model.fit_transform(topic_weights)

# Calculate the dominant topic for each document
dominant_topic = topic_weights.argmax(axis=1)
merged_df['Dominant_Topic'] = dominant_topic

# Define topics and their corresponding colors
topics = ['Topic 1: COVID-19 impact on daily necessities',
          'Topic 2: Grocery shopping during COVID-19',
          'Topic 3: COVID-19 pandemic and consumer behavior',
          'Topic 4: COVID-19 impact on store workers and retail',
          'Topic 5: COVID-19 and food-related concerns']
topic_colors = ['red', 'green', 'blue', 'orange', 'purple']

# Plot the t-SNE visualization
plt.figure(figsize=(10, 8))
for i in range(number_topics):
    plt.scatter(tsne_data[:, 0][merged_df['Dominant_Topic'] == i],
                tsne_data[:, 1][merged_df['Dominant_Topic'] == i],
                color=topic_colors[i],
                alpha=0.7,
                label=topics[i])
plt.legend()
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.title('LDA Topic Visualization')
plt.show()
```

```python
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from textblob import TextBlob
import re
```

```python
def get_sentiment_textblob(text):
    sentiment = TextBlob(text).sentiment.polarity
    if sentiment > 0:
        return "Positive"
    elif sentiment < 0:
        return "Negative"
    else:
        return "Neutral"
```

```python
def get_sent_textblob_cont(text):
    return TextBlob(text).sentiment.polarity
```

```python
def get_sentiment_vader(text):
    sia = SentimentIntensityAnalyzer()
    sentiment = sia.polarity_scores(text)
    
    if sentiment['compound'] > 0.05:
        return "Positive"
    elif sentiment['compound'] < -0.05:
        return "Negative"
    else:
        return "Neutral"
```

```python
def get_sent_vader_cont(text):
    sia = SentimentIntensityAnalyzer()
    return sia.polarity_scores(text)['compound']
```

```python
merged_df['sentiment_textblob'] = merged_df['Text_Cleaned'].apply(get_sentiment_textblob)
merged_df['sent_textblob_cont'] = merged_df['Text_Cleaned'].apply(get_sent_textblob_cont)
merged_df['sentiment_vader'] = merged_df['Text_Cleaned'].apply(get_sentiment_vader)
merged_df['sent_vader_cont'] = merged_df['Text_Cleaned'].apply(get_sent_vader_cont)
```

The reason for this is that there is the feelings of optimism and delight that people (both students and workers) expressed about staying home and not needing to go to school or work. This trend can be seen in around mid to late March when the sentiment scores for vader are the highest.
```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
import matplotlib.pyplot as plt
import seaborn as sns
from nrclex import NRCLex
import matplotlib.dates as mdates
```

```python
# Function to get the top emotion from a text
def get_emotion_from_text(text):
    emotion = NRCLex(text)
    return emotion.top_emotions[0][0]

# Add emotion detection to the dataframe
merged_df['Emotion'] = merged_df['Text_Cleaned'].apply(get_emotion_from_text)

# Print the first few rows of the merged dataframe
merged_df
```

```python
# Count the frequency of each unique value in the 'Emotion' column
emotion_counts = merged_df['Emotion'].value_counts()

# Create a bar plot for emotion frequency distribution
plt.figure(figsize=(10,6))
sns.barplot(x=emotion_counts.index, y=emotion_counts.values, alpha=0.8)
plt.title('Emotion Frequency Distribution')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('Emotion', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

merged_df['TweetAt'] = pd.to_datetime(merged_df['TweetAt'])

emotion_over_time = filtered_df.groupby([filtered_df['TweetAt'].dt.date, 'Emotion']).size().reset_index(name='Counts')

emotion_over_time_pivot = emotion_over_time.pivot(index='TweetAt', columns='Emotion', values='Counts').fillna(0)

fig, ax = plt.subplots()
emotion_over_time_pivot.plot(kind='line', stacked=False, ax=ax)

ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval=1))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))

plt.setp(ax.get_xticklabels(), rotation=45, horizontalalignment='right')

plt.xlabel('Date')
plt.ylabel('Frequency')
plt.title('Emotion Distribution Over Time')
plt.tight_layout()
plt.show()
```
